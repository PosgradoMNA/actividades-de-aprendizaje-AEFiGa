### Solucion 1:

#### Solución 1.1:  Descartar las observaciones con valores faltantes

```python
df1_1 = df.copy()
df1_1.dropna(inplace = True)
df1_1
```

La solución me parece aceptable porque se eliminaron 42 filas de 30,000. Esto equivale a 0.14% de los datos, lo cual no es significativo.

#### Solucion 1.2: Descartar las columnas donde falte al menos un elemento

```python
df1_2 = df.copy()
df1_2.dropna(axis = 1, inplace = True) # axis 1 is columns / axis 0 is rows. 
df1_2
```

Esta solución no es buena, ya que como se vio en el análisis inicial todas las columnas presentan datos faltantes.

#### Solucion 1.3: Descartar filas donde falten todos los valores

```python
df1_3 = df.copy()
df1_3.dropna(how='all', inplace = True)
df1_3
```

Esta solución tampoco es buena, ya que no logra eliminar niguna de las filas con valores faltantes.

#### Solucion 1.4: Descartar filas con un Threshold

Mantenga solo las filas con al menos 20 valores que  **NO SEAN** `nan`

```python
df1_4 = df.copy()
df1_4.dropna(thresh=20, inplace = True) # In a row, it needs at least 4 nan values is needed, to maintain in df
df1_4 # in case of column  add   axis=1 
```

```python
df1_4.isnull().any()
```

Es una solución en la que muchas columnas siguen mostrando valores faltantes. Revisar si se puede ajustar el threshold para obtener mejores resultados.

#### Solucion 1.5: Descartar columnas con un Threshold

```python
df1_5 = df.copy()
df1_5.dropna(thresh = 29995,  #if there is not 29995 nan values, the column will be eliminated 
           axis = 1, 
           inplace = True
           ) 
df1_5 
```

No es una solución viable porque elimina una gran cantidad de columnas. Revisar si se puede ajustar el threshold para obtener mejores resultados.

# Crear un informe de limpieza de datos


Registrar sus actividades de limpieza de datos es esencial para registrar las modificaciones de los datos. 

Los futuros proyectos de minería de datos se beneficiarán de los detalles del trabajo disponible.


Es una excelente idea considerar las siguientes cuestiones cuando genere el informe:


* ¿Qué tipos de ruido se han producido en los datos?
* ¿Qué métodos utiliza para eliminar el ruido? 
    
    > ¿Qué técnicas han demostrado ser eficaces?

* ¿Existen casos o atributos que no se pueden recuperar? 
> Asegúrese de registrar los datos que se han excluido por causas del ruido.
