### Data Normalization in Python

Es una técnica que permite que variables en diferentes rango puedan tener el mismo impacto, sin importar la magnitud del valor comparado con otro.

Existen distintas formas de normalizas. Las tres que se verán en este curso son:
  - Valor escalable sencillo (valores entre 0 y 1)
  $$x_{new}=\frac{x_{old}}{x_{max}}$$
  - Min-Max (valores entre 0 y 1)
  $$x_{new}=\frac{x_{old}-x_{min}}{x_{max}-x_{min}}$$
  - Z score (valores alrededor de 0, pero con valores típicos de -3 y +3)
   $$x_{new}=\frac{x_{old}-\mu}{\sigma}$$

```python
df3 = df.copy()
df3["length"] = df3["length"] / df3["length"].max()
df3["length"]
```

```python
df3 = df.copy()
df3["length"] = (df3["length"] - df3["length"].min()) / (df3["length"].max() - df3["length"].min())
df3["length"]
```

```python
df3 = df.copy()
df3["length"] = (df3["length"] - df3["length"].mean()) / df3["length"].std()
df3["length"]
```

### Linear Regression and Multiple Linear Regression

Simple Linear Regression (SLR): Existen dos variables, una es el predictor y el otro el objetivo. Queremos saber el comportamiento lineal de la relación entre ambas variables.
Dada la ecuación $y=b_0 +b_1x$, tenemos los parámetros:
 - $b_0$: La intercepción
 - $b_1$: La pendiente

Los pasos a seguir son:
 1. Importar linear_model de scikit-learn
  `from sklearn.linear_model import LinearRegression`
 2. Crear un onjeto de regresión linear
  `lm = LinearRegression()`

```python
from sklearn.linear_model import LinearRegression
X = df2[["highway-mpg"]]
Y = df2[["price"]]

lm = LinearRegression()

lm.fit(X,Y)

Yhat = lm.predict(X)

b0 = lm.intercept_ #b0
b1 = lm.coef_ #b1

print("The predicted equation would be y = {}*x + {}".format(b1[0][0], b0[0]))
```

Múltiple Linear Regression (MLR): Se utiliza para explicar la relación entre una salida objetivo y dos o más variables que ayuden a predecir la primera.

Considere la función:
$y = b_0 + b_1x_1 + b_2x_2+\dots b_nx_n$

```python
Z = df2[["horsepower", "curb-weight", "engine-size", "highway-mpg"]]
Y = df2[["price"]]

lm = LinearRegression()

lm.fit(Z,Y)

Yhat = lm.predict(Z)

b0 = lm.intercept_[0]
b1 = lm.coef_[0][0]
b2 = lm.coef_[0][1]
b3 = lm.coef_[0][2]
b4 = lm.coef_[0][3]

print("The predicted equation would be y = {}*x1 + {}*x2 + {}*x3 + {}*x4 + {}".format(b1, b2, b3, b4, b0))
```

### Polynomial Regression and Pipelines

Si un modelo lineal no tiene el ajuste que requerimos, podemos utilizar una regresión polinomial. Esto sirve cuando tenemos una relación curvilínea que resulta de tener términos que tiene alguna relación cuadrática o de otro orden superior.

En este sentido, podemo tener polinomios cuadráticos (2do orden), cúbicos (3er orden) o de orden superior. El grado de la regresión puede hacer una gran diferencia y con el valor estimado.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

pr = PolynomialFeatures(degree=2, include_bias=False)
x_polly = pr.fit_transform(df2[["horsepower", "curb-weight"]])

x_polly
```

```python
SCALE = StandardScaler()
SCALE.fit(df2[["horsepower", "highway-mpg"]])
x_scale = SCALE.transform(df2[["horsepower", "highway-mpg"]])

x_scale
```

Pipelines

Los pasos para poder llegar a una predicción son:

Normalización -> Transformación polinomial -> Regresión lineal

El Pipeline se encarga de las transformaciones

```python
df2["horsepower"].replace('?', '0', inplace=True)
df2["horsepower"] = df2["horsepower"].astype("int")
df2["curb-weight"].replace('?', '0', inplace=True)
df2["curb-weight"] = df2["curb-weight"].astype("int")
df2["engine-size"].replace('?', '0', inplace=True)
df2["engine-size"] = df2["engine-size"].astype("int")
df2["highway-mpg"].replace('?', '0', inplace=True)
df2["highway-mpg"] = df2["highway-mpg"].astype("int")

# df2["horsepower"].unique()
df2.info()
```

```python
from sklearn. preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn. preprocessing import StandardScaler

from sklearn.pipeline import Pipeline

Input = [
        ("scale", StandardScaler()),
        ("polynomial", PolynomialFeatures(degree=2)),
        ("model", LinearRegression())]

pipe = Pipeline(Input)

pipe.fit(df2[["horsepower", "curb-weight", "engine-size", "highway-mpg"]], df2[["price"]])

Y_predict_simple_fit = pipe.predict(df2[["horsepower", "curb-weight", "engine-size", "highway-mpg"]])

print(pipe.score(df2[["horsepower", "curb-weight", "engine-size", "highway-mpg"]], df2[["price"]]))
```

### Model Evaluation

La evaluación de un modelo nos dice la forma en que nuestro modelo se comporta con los datos dados. No nos da información de la precisión del modelo para datos nuevos, pero una forma de arreglar esto es dividir los datos entre entrenamiento, validación y pruebas.

Una forma típida de separar es 70% train, 30% test

```python
from sklearn. model_selection import train_test_split

X = df2[["highway-mpg"]]
Y = df2[["price"]]

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

A una mayor cantidad de datos de entrenamiento, mejor presisión tendrá el modelo, pero menor repetitivilidad.

A una mayor cantidad de datos de validación, menor presisión tendrá el modelo, pero mayor repetitivilidad.

Una de las métricas más comunes de evaluación para modelos es Cross-Validation.

El Dataset se divide en $k$ grupos de igual cantidad. Estos grupos (folds) se van intercalando en pruebas para usar todos menos uno para entrenar y el grupo restante para validar. El proceso se repite hasta que todos los grupos fueron utlizados para entrenar y validar.

Una forma de aplicar Cross-Validation es con `cross_val_score()`

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lm, X, Y, cv=3)

scores.mean()
```

```python
from sklearn.model_selection import cross_val_predict

yhat = cross_val_predict(lm, X,Y,cv=3)
yhat
```

<a id='Module5_3'></a>
### Overfitting, Underfitting and Model Selection

En la selección un grado para una regresión polinomial puede suceder que se seleccione un modelo que acierte en los datos que se le dan al modelo, pero termina absorbiendo el ruido en lugar de sólo los datos.

Esto se puede comprobar con una gráfica que muestre $R^2$ a distintos grados del polinomio.

```python
R_squ = []
order = [1,2,3,4,5,6]

lr = LinearRegression()

for n in order:
    pr = PolynomialFeatures(degree=n)
    x_train_pr = pr.fit_transform(x_train)
    x_test_pr  = pr.fit_transform(x_test)
    lr.fit(x_train_pr, y_train)
    R_squ.append(lr.score(x_test_pr,y_test))
        
print(R_squ)

plt.plot(order,R_squ)
plt.xlabel("Polynomial degree")
plt.ylabel("$R^2$")
plt.title("Ploynomial degree vs $R^2$")
```

### Grid Search

Sklaern tiene una forma de iterar de forma automática por todos los hiperparámetros que pueden reuqerirse utilizando una forma de Validación Cruzada llamada Grid Search.

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

parameter1 = [{"alpha":[0.001, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000], "normalize":[True,False]}]

RR = Ridge()

Grid1 = GridSearchCV(RR, parameter1, cv=4)
Grid1.fit(df2[["horsepower", "curb-weight", "engine-size", "highway-mpg"]], df2[["price"]])
Grid1.best_estimator_

scores = Grid1.cv_results_
scores["mean_test_score"]
```
